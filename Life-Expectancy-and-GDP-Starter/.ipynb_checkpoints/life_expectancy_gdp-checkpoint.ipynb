```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels as sm
```

```python
data = pd.read_csv('all_data.csv')
```

```python
print(data.head())
print(data.dtypes)
print(data.Country.unique())
```

```python
data["GDP_1000"] = data.GDP / 1000
plt.figure(figsize = (10, 8))
sns.barplot(x = 'Country', y = 'GDP', data = data)
plt.xticks(rotation = 30, fontsize = 10)
plt.show()
plt.clf()

sns.scatterplot(x = 'Life expectancy at birth (years)', y = 'GDP', data = data, hue = 'Country')
plt.show()
plt.clf()
```

```python
plt.figure(figsize = (25,20))
plt.subplot(231)
Country_X = data[data.Country == "Chile"]
#ax.set_xticks(Country_Chile["Year"])
sns.scatterplot(x ="Life expectancy at birth (years)", y = "GDP", data = Country_X)
plt.legend(Country_X.Country)
plt.show()
plt.clf()

plt.subplot(232)
Country_X = data[data.Country == "China"]
#ax.set_xticks(Country_Chile["Year"])
sns.scatterplot(x ="Life expectancy at birth (years)", y = "GDP", data = Country_X)
plt.legend(Country_X.Country)
plt.show()
plt.clf()

plt.subplot(233)
Country_X = data[data.Country == "Germany"]
#ax.set_xticks(Country_Chile["Year"])
sns.scatterplot(x ="Life expectancy at birth (years)", y = "GDP", data = Country_X)
plt.legend(Country_X.Country)
plt.show()
plt.clf()

plt.subplot(234)
Country_X = data[data.Country == "Mexico"]
#ax.set_xticks(Country_Chile["Year"])
sns.scatterplot(x ="Life expectancy at birth (years)", y = "GDP", data = Country_X)
plt.legend(Country_X.Country)
plt.show()
plt.clf()

plt.subplot(235)
Country_X = data[data.Country == "United States of America"]
#ax.set_xticks(Country_Chile["Year"])
sns.scatterplot(x ="Life expectancy at birth (years)", y = "GDP", data = Country_X)
plt.legend(Country_X.Country)
plt.show()
plt.clf()

plt.subplot(236)
Country_X = data[data.Country == "Zimbabwe"]
#ax.set_xticks(Country_Chile["Year"])
sns.scatterplot(x ="Life expectancy at birth (years)", y = "GDP", data = Country_X)
plt.legend(Country_X.Country)
plt.show()
```
